5. <b>Carga y Exploración de los Datos: </b> <br>

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot

# Carga de los 3 conjuntos de datos utilizando Pandas.
df1 = pd.read_csv("grupo1.csv")
df2 = pd.read_csv("grupo2.csv")
df3 = pd.read_csv("grupo3.csv")
 
# Realizar una exploración inicial de los datos (descripción estadística y visualización).

# Descripción estadística y visualizacipon datos del grupo 1.
print ("GRUPO 1 - HEAD")
print (df1.head())
print ("")
print ("GRUPO 1 - DESCRIPCION")
print (df1.describe())
print ("")
print (df1)

# Histograma de calificaciones finales del grupo 1.
plt.pyplot.hist(df1.Calificacion, bins=7, alpha=0.5)
plt.pyplot.title('Histograma de Calificaciones - GRUPO 1 ')
plt.pyplot.xlabel('Calificación Final')
plt.pyplot.ylabel('Frecuencia')
```

```python
# Descripción estadística y visualizacipon datos del grupo 2.
print ("GRUPO 2 - HEAD")
print (df2.head())
print ("")
print ("GRUPO 2 - DESCRIPCION")
print (df2.describe())
print ("")
print (df2)

# Histograma de calificaciones finales del grupo 2.
plt.pyplot.hist(df2.Calificacion, bins=8, alpha=0.5)
plt.pyplot.title('Histograma de Calificaciones - GRUPO 2 ')
plt.pyplot.xlabel('Calificación Final')
plt.pyplot.ylabel('Frecuencia')
```

```python
# Descripción estadística y visualizacipon datos del grupo 3.
print ("GRUPO 3 - HEAD")
print (df3.head())
print ("")
print ("GRUPO 3 - DESCRIPCION")
print (df3.describe())
print ("")
print (df3)

# Histograma de calificaciones finales del grupo 3.
plt.pyplot.hist(df3.Calificacion, bins=5, alpha=0.5)
plt.pyplot.title('Histograma de Calificaciones - GRUPO 3 ')
plt.pyplot.xlabel('Calificación Final')
plt.pyplot.ylabel('Frecuencia')
```

6. <b> Modelo de Regresión Lineal: </b> <br>
Entrenar un modelo de regresión lineal simple usando scikit-learn o statsmodels. <br>
Obtener la ecuación de la línea de regresión (pendiente e intercepto). <br>
Calcular el coeficiente de correlación de Pearson entre las horas de estudio y las calificaciones.<br> 

7. <b> Visualización de Resultados: </b> <br>
Graficar los datos junto con la línea de regresión. <br>
Incluir un gráfico de dispersión (scatter plot) y la línea de regresión.<br>

```python
from sklearn.linear_model import LinearRegression # Modelo de regresión Lineal
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

horasEstudio = df1[['HorasEstudio']]
calificacion = df1[['Calificacion']]

# Crear y entrenar el modelo de regresión Lineal para el df1 (grupo 1).
modelo = LinearRegression()
modelo.fit(horasEstudio, calificacion)

# Coeficientes de La regresion del grupo 1.
print("Coeficiente de la pendiente (m):", modelo.coef_[0])
print("Intercepcion de la recta (b):", modelo.intercept_)

# Datos de entrada para el coeficiente de correlación del grupo 1.
horasEstudio = df1['HorasEstudio'].values
calificacion = df1['Calificacion'].values

# Calcular el coeficiente de correlacion de Pearson del grupo 1.
coeficiente_correlacion, g1_valor = pearsonr (horasEstudio, calificacion)

print ("Coeficiente de correlacion de Pearson: ", coeficiente_correlacion)
print ("Valor g1: ", g1_valor)


# Visualización de los resultados.

sns.scatterplot(x = horasEstudio , y = calificacion)
plt.title( 'Diagrama de Dispersión de Horas de Estudio vs. Calificación Final')
plt.xlabel( 'Calificación Final')
plt.ylabel( 'Horas de Estudio')
plt.grid(True)
plt.show()

sns.set(style="whitegrid")
sns.regplot(x='HorasEstudio', y='Calificacion', data=df1, scatter=True, color='blue')
plt.title('Diagrama de Dispersión de Estudio vs Calificación Final')
plt.xlabel('Horas Estudiadas')
plt.ylabel('Calificación')
plt.grid(True)
plt.show()
```

```python
# Para el grupo 2:

horasEstudio = df2[['HorasEstudio']]
calificacion = df2[['Calificacion']]

# Crear y entrenar el modelo de regresión Lineal para el df2 (grupo 2).
modelo = LinearRegression()
modelo.fit(horasEstudio, calificacion)

# Coeficientes de La regresion del grupo 2.
print("Coeficiente de la pendiente (m):", modelo.coef_[0])
print("Intercepcion de la recta (b):", modelo.intercept_)

# Datos de entrada para el coeficiente de correlación del grupo 2.
horasEstudio = df2['HorasEstudio'].values
calificacion = df2['Calificacion'].values

# Calcular el coeficiente de correlacion de Pearson del grupo 2.
coeficiente_correlacion, g2_valor = pearsonr (horasEstudio, calificacion)

print ("Coeficiente de correlacion de Pearson: ", coeficiente_correlacion)
print ("Valor g2: ", g2_valor)


# Visualización de los resultados.

sns.scatterplot(x = horasEstudio , y = calificacion)
plt.title( 'Diagrama de Dispersión de Horas de Estudio vs. Calificación Final')
plt.xlabel( 'Calificación Final')
plt.ylabel( 'Horas de Estudio')
plt.grid(True)
plt.show()

sns.set(style="whitegrid")
sns.regplot(x='HorasEstudio', y='Calificacion', data=df2, scatter=True, color='blue')
plt.title('Diagrama de Dispersión de Estudio vs Calificación Final')
plt.xlabel('Horas Estudiadas')
plt.ylabel('Calificación')
plt.grid(True)
plt.show()
```

```python
# Para el grupo 3.
horasEstudio = df3[['HorasEstudio']]
calificacion = df3[['Calificacion']]

# Crear y entrenar el modelo de regresión Lineal para el df3 (grupo 3).
modelo = LinearRegression()
modelo.fit(horasEstudio, calificacion)

# Coeficientes de La regresion del grupo 3.
print("Coeficiente de la pendiente (m):", modelo.coef_[0])
print("Intercepcion de la recta (b):", modelo.intercept_)

# Datos de entrada para el coeficiente de correlación del grupo 3.
horasEstudio = df3['HorasEstudio'].values
calificacion = df3['Calificacion'].values

# Calcular el coeficiente de correlacion de Pearson del grupo 3.
coeficiente_correlacion, g3_valor = pearsonr (horasEstudio, calificacion)

print ("Coeficiente de correlacion de Pearson: ", coeficiente_correlacion)
print ("Valor g1: ", g3_valor)


# Visualización de los resultados.

sns.scatterplot(x = horasEstudio , y = calificacion)
plt.title( 'Diagrama de Dispersión de Horas de Estudio vs. Calificación Final')
plt.xlabel( 'Calificación Final')
plt.ylabel( 'Horas de Estudio')
plt.grid(True)
plt.show()

sns.set(style="whitegrid")
sns.regplot(x='HorasEstudio', y='Calificacion', data=df3, scatter=True, color='blue')
plt.title('Diagrama de Dispersión de Estudio vs Calificación Final')
plt.xlabel('Horas Estudiadas')
plt.ylabel('Calificación')
plt.grid(True)
plt.show()
```
