```python
import math
from statistics import *
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
autos = pd.read_csv("autos2.csv")
autos
```

1. Estadisticas Descriptivas Basicas y Medidas de Tendencia Central 

```python
print ("El valor minimo de los precios: " , autos.price.min())
print ("El valor maximo de los precios: ", autos.price.max())
moda = multimode(autos.price)
print ("La moda de los precios: " , moda)
print ("La media de los precios: " , autos.price.mean())
print ("La mediana de los precios: " , autos.price.median())
autos.describe()
```

```python
n = len(autos)
k = int (1+ math.log2(n))

print ("Valor de n=", n, "\nK o numero de intervalos o clases=", k) 
```

```python
plt.pyplot.hist(autos.price, bins=8)
plt.pyplot.title("Histograma de Precios de los Automóviles")
plt.pyplot.xlabel("Precio del Automóvil")
plt.pyplot.ylabel("Frecuencia ")
```

El histograma anterior podemos interpretarlo como que el rango de los precios de los autos más vendidos oscila entre los 5000 y 10,000 dlls. y que conforme el precio de los mismos va elevando, podemos notar una variación que disminuye y que se podria relacionar con lo costoso que es adquirir un auto de 45,000 dlls.

```python
sns.boxplot(x = "body-style", y = "price", data=autos)
plt.pyplot.title("Relación Estadistica del precio del automóvil y el tipo de vehículo")
plt.pyplot.xlabel("Estilo del Cuerpo del Automóvil")
plt.pyplot.ylabel("Precio del Automóvil")
```

La varianza de los precios de los automóviles oscila conforme a su estilo, aunque podría influir el tipo de uso que le quiera dar quien adquiera el mismo, las estadísticas reflejan que los autos más costosos son los covertible y el hard top, aunque en las otras 3 clasificaciones podemos observar datos atípicos en cuanto a la varianza del precio.

```python
sns.boxplot(x = "drive-wheels", y = "price", data=autos)
plt.pyplot.title("Relación Estadistica del precio del automóvil y el tipo de tracción de ruedas del vehículo.")
plt.pyplot.xlabel("Tipo de Tracción")
plt.pyplot.ylabel("Precio del Automóvil")
```

En este gráfico podemos observar que los vehículos 4wd  y fwd oscilan entre los precios de 5,000 apróximadamente hasta los 15,000, mientras que los vehículos rwd son los que con más frecuencia se presentan y que a mayor precio se encuentran, aunque podemos observar puntos atípicos en cada una de las clasificaciones del tipo de tracción. 

```python
sns.boxplot(x = "engine-location", y = "price", data=autos)
plt.pyplot.title("Relación Estadistica del precio del automóvil y la ubicación del motor.")
plt.pyplot.xlabel("Ubicación del motor")
plt.pyplot.ylabel("Precio del Automóvil")
```

En base a este último gráfico relacionado a la ubicación del motor y el precio del automóvil podemos concluir que en los autos que su precio oscila entre los 5,000 y 20,000 el motor se encuentra al frente, teniendo algunos resultados o datos atípicos, sin embargo podemos observar pues que los coches más costosos son aquellos en los cuales su motor se encuentra en la parte trasera.
