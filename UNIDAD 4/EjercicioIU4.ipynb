Ejercicio 1 
Una revison de los pesos de 500 estudiantes del ITD, evidencio que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

Generar aleatoriamente las edades de los 500 estudiantes con una media de 70 kg. y una desviacion estandar de 4 kg.
Realizar un histograma de las edades.
Graficar la funcion de densisad de probabilidad.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generar datos para la distribucion normal

media = 70 #Media
desv_std = 4 # Desviacion estandar
num_datos = 500

#Generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] #Mostrar los primeros 50
```

```python
# Crear un rango de valores para crear una densidad
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

# Calcular la PDF de la distribucion normal
pdf = 1 / (desv_std * np.sqrt (2*np.pi))* np.exp(-(x-media)**2 / (2 * desv_std **2))

#Para no imprimir toda la lista de valores, desplegamos los primeros 50
print ("Imprimir los primeros 50 valores de x", x[0:50], ", y los primeros 50 de pdf", pdf[0:50])
```

```python
#Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b') #Histograma de los datos 
plt.title('Distribucion Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
#Visualizar PDF
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Funcion de Densidad de Probabilidad (PDF) - Distribucion Normal')
plt.xlabel('Pesos en Kg. de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar histograma
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot( x, pdf, 'k', linewidth=2, )

plt.title('Distribucion Normal')
plt.xlabel('Peso en kg de los estudiantes ' )
plt.grid(True)
plt.show()
```

Ejercicio 2. 
Una revision de los pesos de 500 estudiantes del ITD, evidencio que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:
Cuantos estudiantes pesan menos de 75 kg.
Cuantos estudiantes pesan menos de 62 kg.
Cuantos estudiantes pesan mas de 79 kg.

Primero necesitamos calcular la probabilida de que un estudiante pese menos de 75 kg utilizando la funcion de distribucion acumulativa (CDF) de la distribucion normal

```python
plt.figure( figsize=(6,4))

# Parametros de la distribucion normal
media = 70 # Media
desv_std = 4 # Desviacion estandar
total_estudiantes=500

# Variable para calculo de la probabilidad de que un estidnate pese menos de 75 kg.
peso_limite = 75

# Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75 kg
# utilizando la funcion de distribucion acumulativa (CDF) de la distribucion normal
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)
```

```python
# Calculo del numero de estudiantes que pesan menos de 75 kg
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print ("Probabilidad de que un estudiante pese menos de 75 kg: " , prob_menos_75kg)
print ("Numero de estudiantes que pesan menos de 75 kg: " , estudiantes_menos_75kg)

# Graficar la distribucion normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot (x, y, 'b', linewidth=2)
plt.fill_between(x, y, where =  (x <= peso_limite), color='skyblue', alpha=0.5)
plt.title(r'Probabilidad de que un estudiante pese menos de 75 kg, \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure( figsize=(6,4))

# Parametros de la distribucion normal
media = 70 # Media
desv_std = 4 # Desviacion estandar
total_estudiantes=500

# Variable para calculo de la probabilidad de que un estidnate pese menos de 62 kg.
peso_limite = 62

# Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 62 kg
# utilizando la funcion de distribucion acumulativa (CDF) de la distribucion normal
prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)

# Calculo del numero de estudiantes que pesan menos de 62 kg
estudiantes_menos_62kg = total_estudiantes * prob_menos_62kg

print ("Probabilidad de que un estudiante pese menos de 62 kg: " , prob_menos_62kg)
print ("Numero de estudiantes que pesan menos de 62 kg: " , estudiantes_menos_62kg)

# Calcular la distribucion normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Graficar la distribucion normal
plt.plot (x, y, 'b', linewidth=2)
plt.fill_between(x, y, where =  (x <= peso_limite), color='skyblue', alpha=0.5)
plt.title(r'Probabilidad de que un estudiante pese menos de 62 kg, \n $\mu=70$, $\sigma=4$, x=500')

plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure( figsize=(6,4))

# Parametros de la distribucion normal
media = 70 # Media
desv_std = 4 # Desviacion estandar
peso_limite = 79 # Peso limite
total_estudiantes=500

# Calculamos la probabilidad de que un estudiante pese menos o igual a 79 kg
prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)

# La probabilidad de que un estudiante pese mas de 79 kg es complementaria
prob_mas_79kg = 1- prob_menos_79kg

# Calculamos el numero aproximado de estudiantes que pesan mas de 79 kg
estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg

print ("Probabilidad de que un estudiante pese mas de 79 kg: " , prob_mas_79kg)
print ("Numero de estudiantes que pesan mas de 79 kg: " , estudiantes_mas_79kg)

# Calcular la distribucion normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Graficar la distribucion normal
plt.plot (x, y, 'b', linewidth=2)
plt.fill_between(x, y, where =  (x >= peso_limite), color='skyblue', alpha=0.5)
plt.title(r'Probabilidad de que un estudiante pese mas de 79 kg, \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

## Ejercicio 3
Realizar un programa en Python que resuelva el ejercicio:
Una revision de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

La probabilidad de estudiantes que pesan mas de 68 y menos de 72 kg.
Cuantos estudiantes pesan mas de 68 y menos de 72 kg.

```python
# Parametros de la distribucion normal
media = 70 # Media
desv_std = 4 # Diacion estandar
total_estudiantes=500

# Variable para calculo de la probabilidad de que un estudiante pese menos de 72 kg.
peso_limite = 72

prob_menos_72kg = norm.cdf(peso_limite, media, desv_std)

# Variable para calculo de la probabilidad de que un estudiante pese mas de 68 kg.
peso_minimo = 68

prob_mas_68kg = norm.cdf(peso_minimo, media, desv_std)

probabilidad_mayor68_menor72 = prob_menos_72kg-prob_mas_68kg

# Calculo del numero de estudiantes
estudiantes_mas68_y_menos_72 = total_estudiantes * probabilidad_mayor68_menor72

print ("Probabilidad de que un estudiante pese mas de 68 kg y menos de 72 kg: " , probabilidad_mayor68_menor72)
print ("Numero de estudiantes que pesan menos de 68 kg y menos de 72 kg kg: " , estudiantes_mas68_y_menos_72 )

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Graficar la distribucion normal
plt.plot (x, y, 'b', linewidth=2)
plt.fill_between(x, y, where =  ((x >= peso_minimo)& (x<=peso_limite)), color='skyblue', alpha=0.5)
plt.title(r'Probabilidad de que un estudiante pese mas de 68 kg y menos de 72 kg, \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
